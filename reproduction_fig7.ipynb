In this notebook we describe how to generate fig 7 from the paper.

```python
import math
import numpy as np
from pylab import *
import os
import os.path
import json
import glycogen_module as gm
import random
import pandas as pd
import matplotlib.pyplot as plt

random.seed(123)
```

## Generating data

As a example, we will generate the case of a branching mechanism with fingerprints value \{6,4,4\} (green curves on the top left panel of Fig7), with $\Gamma = 0.2$

For that purpose, we set `parameters['GS'] = 0.2` and `parameters['GBE'] = 1.0`

Again, by default, we generate a small number of simulations, with low molecular weight (N = 2500), to keep the computational time low.

```python
with open("parameters.json", "r") as f:
    parameters = json.load(f)

parameters['size_spec_gbe_spacing'] = 6
parameters['size_spec_gbe_leftover'] = 4
parameters['size_spec_gbe_transferred'] = 4
parameters['b'] = 5.4
parameters['GS'] = 0.2
parameters['GBE'] = 1.0
parameters['number_of_glucose_to_be_fixed'] = 1500

Number_of_simulations = 20

path_of_local_folder = os.getcwd()
folder_name = '/data_for_fig7'
os.mkdir(path_of_local_folder+folder_name)
data_path = path_of_local_folder+folder_name
```

```python
from algorithm import Gillespie_step
```

```python
C0 = {"GS": parameters['GS'], "GBE": parameters['GBE'], "GP": 0.0, "GDE": 0.0}

identifier = str(parameters['size_spec_gbe_spacing'])+str(parameters['size_spec_gbe_leftover']) + \
    str(parameters['size_spec_gbe_transferred']) + \
    '_'+str(parameters['GS']/parameters['GBE'])


os.mkdir(data_path+'/'+identifier)
folder_path = data_path+'/'+identifier

for i in range(Number_of_simulations):

    my_granule = gm.glycogen_structure(parameters, 2, 1)
    N = my_granule.number_of_glucose_fixed()

    while N < parameters['number_of_glucose_to_be_fixed']:
        # if N%5000 == 0:
        #    print('avancement = ', 100*N/parameters['number_of_glucose_to_be_fixed'], '%','   part 1/2')

        reaction, time = Gillespie_step(my_granule, C0)
        # print(reaction)
        if reaction == 'Act_gs()':
            my_granule.Act_gs()
        elif reaction == 'Act_gp()':
            my_granule.Act_gp()
        elif reaction == 'Act_gbe()':
            my_granule.Act_gbe_flexible_model()
        elif reaction == 'Act_gde()':
            my_granule.Act_gde()

        N = my_granule.number_of_glucose_fixed()

    name = 'glycogen_structure'+str(len(os.listdir(folder_path))+1)+'.json'
    my_granule.write_structure(folder_path+'/'+name)
```

```python
def cld(dict):
    histo = []
    for chain in dict.values():
        histo.append(len(chain['glucose_location']))
    return histo


def cumulative_cld(path_of_folder: str) -> array:
    cld_cumulated = []

    for i in range(len(os.listdir(path_of_folder))):
        file_name = 'glycogen_structure'+str(i+1)+'.json'
        path = path_of_folder+'/'+file_name

        with open(path, "r") as f:
            my_granule_data = json.load(f)

        distrib = cld(my_granule_data)
        cld_cumulated = cld_cumulated + distrib

    return (cld_cumulated)


def plot_cld_from_distrib(liste):

    bins1 = np.linspace(0, max(liste), max(liste)+1)
    plt.figure(figsize=(14, 7))
    plt.hist(liste, bins=bins1, color='grey', alpha=0.5,
             edgecolor='black', label='packed', density=1)
    plt.xlabel('DP', fontsize=20)
    plt.ylabel('relative abundances', fontsize=20)
    plt.legend()
    plt.tick_params(axis='both', which='major', labelsize=17)
    plt.xlim([0, 50])
    plt.show()


def cumulative_cld(path_of_folder: str) -> array:
    cld_cumulated = []

    for i in range(len(os.listdir(path_of_folder))):
        file_name = 'glycogen_structure'+str(i+1)+'.json'
        path = path_of_folder+'/'+file_name

        with open(path, "r") as f:
            my_granule_data = json.load(f)

        distrib = cld(my_granule_data)
        cld_cumulated = cld_cumulated + distrib

    return (cld_cumulated)


def hist_into_curve(distrib: list):
    'this function takes an histogram list and return X and Y arrays to plot the associated curve'

    total = len(distrib)
    maximum = max(distrib)
    # print(maximum)
    X = np.linspace(0, maximum, maximum+1)
    Y = []
    for k in X:
        counter = 0
        for i in range(len(distrib)):
            if k == distrib[i]:
                counter += 1
        Y.append(counter)

    Y = np.asarray(Y)
    return X, Y*1.0/total
```
